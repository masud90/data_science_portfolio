## Data preprocessing

```python
# Convert categorical columns to numeric using one-hot encoding
categorical_cols = ['Incidence', 'Brand', 'Last_Inc_Brand', 'Promotion_1', 'Promotion_2', 'Promotion_3', 'Promotion_4', 'Promotion_5', 'Sex', 'Marital status', 'Education', 'Occupation', 'Settlement size']
purchase_data_encoded = pd.get_dummies(purchase_data, columns=categorical_cols)

# Scale the data
scaler = StandardScaler()
scaled_data = scaler.fit_transform(purchase_data_encoded)

# Convert scaled data back to DataFrame
scaled_df = pd.DataFrame(scaled_data, columns=purchase_data_encoded.columns)
```

## Apply K Means Clustering

```python
# Function to apply K-means and determine the optimal number of clusters
def apply_kmeans(data, n_clusters):
    kmeans = KMeans(n_clusters=n_clusters, random_state=42)
    clusters = kmeans.fit_predict(data)
    return clusters, kmeans.inertia_, silhouette_score(data, clusters)

# Determine the optimal number of clusters using the elbow method and silhouette score
inertia = []
sil_scores = []
K = range(2, 11)
for k in K:
    clusters, inertia_k, sil_score_k = apply_kmeans(scaled_df, k)
    inertia.append(inertia_k)
    sil_scores.append(sil_score_k)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(K, inertia, 'bx-')
plt.xlabel('k')
plt.ylabel('Inertia')
plt.title('Elbow Method for Optimal k')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(K, sil_scores, 'bx-')
plt.xlabel('k')
plt.ylabel('Silhouette Score')
plt.title('Silhouette Score for Optimal k')
plt.show()
```

```python
# Apply K-means with the chosen number of clusters
optimal_k = 4  # Choose based on the elbow method and silhouette score
kmeans_clusters, _, _ = apply_kmeans(scaled_df, optimal_k)
purchase_data['KMeans_Cluster'] = kmeans_clusters
```

## Apply Hierarchical Clustering

```python
# Hierarchical clustering
linked = linkage(scaled_df, 'ward')
plt.figure(figsize=(10, 6))
dendrogram(linked, orientation='top', distance_sort='descending', show_leaf_counts=True)
plt.title('Dendrogram for Hierarchical Clustering')
plt.show()
```

```python
# Apply hierarchical clustering
h_clusters = fcluster(linked, optimal_k, criterion='maxclust')
purchase_data['Hierarchical_Cluster'] = h_clusters
```

## Analyze and Interpret Segments

```python
# Analyze K-means clusters
kmeans_analysis = purchase_data.groupby('KMeans_Cluster').mean()
print("K-means Cluster Analysis:")
print(kmeans_analysis)
```

```python
# Analyze hierarchical clusters
hierarchical_analysis = purchase_data.groupby('Hierarchical_Cluster').mean()
print("Hierarchical Cluster Analysis:")
print(hierarchical_analysis)
```
